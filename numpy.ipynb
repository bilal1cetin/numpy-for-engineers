intergrals and derivatives

```python
x = np.linspace(1,10,100)
y = 1/x**2 * np.sin(x)
dydx = np.gradient(y,x)
y_int = np.cumsum(y) * (x[1]-x[0])
```

```python
plt.plot(x,y)
plt.plot(x,dydx)
plt.plot(x,y_int)
```

# Example <a class="anchor" id="notebook-tips"></a>

Let y = e^(-x/10)sin(x). Consider 10000x intervals in the range[0,10]

1. Plot the function y vs. x
2. compute mean and standard deviation of y for x in [4,7]
3. for x in the range [4,7], find the value y* such that %80 of y values are less than y*
4. plot dy/dx vs. x
5. find the locations where dy/dx = 0 

```python

x  = np.linspace(0,10,10001)
y = np.exp(-x/10) * np.sin(x)
plt.plot(x,y)
```

```python
#2
np.mean(y[(x >= 4)*(x <= 7)])
np.std(y[(x >= 4)*(x <= 7)])
```

```python
#3
np.percentile(y[(x >= 4)*(x <= 7)], 80)
```

```python
#4
plt.plot(x, np.gradient(y,x))
```

```python
#5
dydx = np.gradient(y,x)
```

```python
# multiply consecutive elements to find negative values
x[1:][dydx[1:] * dydx[:-1] < 0]
```

```python
#sum every number from 0 to 10000 expect for those that can ben divided by 4 or 7

number = np.arange(10001)
np.cumsum(number[(number % 4 != 0) * (number % 7 != 0)][:30])[-1]
```
